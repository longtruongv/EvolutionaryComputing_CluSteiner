```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics
```

# MFEA

```python
output_data = pd.read_csv('output/MFEA_2022_11_13_21_46_14_0186.txt', header=None)
output_data.columns = ['generation', 'bf1', 'bf2']
output_data.head()
```

```python
generation_list = output_data['generation'].tolist()
bf1_list = output_data['bf1'].tolist()
bf2_list = output_data['bf2'].tolist()

len(generation_list)
```

```python
bf1_first = bf1_list[0]
bf1_bestfound = bf1_list[-1]

bf1_normalized = [(x - bf1_bestfound) / (bf1_first - bf1_bestfound) for x in bf1_list]
bf1_normalized
```

```python
bf2_first = bf2_list[0]
bf2_bestfound = bf2_list[-1]

bf2_normalized = [(x - bf2_bestfound) / (bf2_first - bf2_bestfound) for x in bf2_list]
bf2_normalized
```

```python
plt.figure(figsize=(20, 10))
plt.plot(generation_list, bf1_normalized, color="blue", alpha=1, label='MFEA eil101')
plt.plot(generation_list, bf2_normalized, color="red", alpha=1, label='MFEA eil51')
leg = plt.legend(loc='upper right')
plt.show()
```

```python
m1 = statistics.mean(bf1_normalized)
m2 = statistics.mean(bf2_normalized)
m1, m2
```
